```python
import numpy as np
import scipy.sparse as spr
import matplotlib.pyplot as plt
import networkx as nx
```

```python
wiki = nx.DiGraph()
```

```python
total_lines_seen = 0
with open("/home/cory/CC21-22/Teaching/B7 - CP341/HW Files/Day 4 - PageRank/links.tsv") as f:
    line = f.readline().strip()
    while total_lines_seen < 200:
        total_lines_seen += 1
        if len(line) > 0 and line[0] != "#":
            u,v = line.split("\t")
            wiki.add_edge(u,v)
        line = f.readline().strip()   
```

```python
nx.draw(wiki)
```

```python
small = nx.DiGraph()
```

```python
small.add_edge(
    "A", "B"
)
small.add_edge(
    "A", "C"
)
small.add_edge(
    "A", "D"
)
small.add_edge(
    "B", "D"
)
small.add_edge(
    "C", "D"
)
small.add_edge(
    "D", "A"
)
```

```python
nx.draw(small)
```

```python
adj = nx.to_scipy_sparse_array(small).T
```

```python
adj.todense()
```

```python
adj.sum(0)
```

```python
S = (adj / adj.sum(0))
```

```python
goog = .85 * S + .15*((1/4)*np.ones_like(S))
```

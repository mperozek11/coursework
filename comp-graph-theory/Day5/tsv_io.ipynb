```python
import numpy as np
import scipy.sparse as spr
import matplotlib.pyplot as plt
import networkx as nx
```

```python
wiki = nx.DiGraph()
```

```python
total_lines_seen = 0
with open('/Users/maxperozek/CP341/Day4/wikispeedia_paths-and-graph/links.tsv') as f:
    line = f.readline()
    while line:
        line = f.readline().strip()

        if total_lines_seen > 200:
            break
        total_lines_seen += 1
        if len(line) > 0 and line[0] != '#':
            u, v = line.split('\t')
            wiki.add_edge(u, v)
```

```python
nx.draw(wiki)
```

```python
small = nx.DiGraph()
```

```python
# Networkx dependency issue
# nx.to_scipy_sparse_array(small).todense().T
adj = nx.to_scipy_sparse_array(small).T
```

```python
adj.sum(0)
```

```python
S = (adj / adj.sum(0))
```

```python
goog = .85 * S + .15 * ((1/4) * np.ones_like(S)) # 4 because there are 4 total websites
```

```python
current = 1/4 * np.ones(4)
```

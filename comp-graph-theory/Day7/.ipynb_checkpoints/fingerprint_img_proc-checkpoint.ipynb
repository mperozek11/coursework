```python
import numpy as np
```

```python
import networkx as nx
```

```python
import sklearn
```

```python
from skimage.io import imread
```

```python
import matplotlib.pyplot as plt
```

```python
im = 1 - (imread("/home/cory/CC21-22/Teaching/B7 - CP341/Daily Work/Day 7/00328ce5.png") / 255)
```

```python
im = im[:,:,0]
```

```python
im.shape
```

```python
plt.matshow(im)
```

```python
im2 = im.copy()
thresh = .5
```

```python
im2[ im2 < thresh] = 0
im2[ im2 >= thresh] = 1.0
```

```python
plt.matshow(im2)
```

```python
from skimage.morphology import skeletonize
```

```python
skel = skeletonize(im2)
```

```python
plt.matshow(skel)
```

```python
nonzero_x, nonzero_y = np.nonzero(skel)
```

```python
nonzero_y.shape
```

```python
coords = np.stack([nonzero_x, nonzero_y]).T
```

```python
coords.shape
```

```python
from sklearn.neighbors import radius_neighbors_graph
```

```python
adj = radius_neighbors_graph(coords, np.sqrt(2))
```

```python
gr = nx.from_scipy_sparse_array(adj)
```

```python
nx.draw(gr, 
        pos=coords,
       node_size=0.0)
```

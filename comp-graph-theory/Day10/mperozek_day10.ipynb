0b) Write a method that converts one of these images to graphs. Your method should take one of these images and produce a graph. Similar to the fingerprint example, your graph should have pixels as nodes, and include an edge wherever two pixels are next to one another. However, here we actually care about the shape - so don't skeletonize the image first.

```python
def im_2_graph(im_mat):
    locs_x, locs_y = np.nonzero(im_mat)
    adj = radius_neighbors_graph(np.stack([locs_x, locs_y]).T, math.sqrt(2))
    return nx.from_numpy_matrix(adj)
```

0c) Write a method that takes in a graph and a set of t values, and produces the estimated Heat Kernel Signature of that graph at those time points. There are several steps to this; refer to the wikipedia article for HKS and to today's slides.

```python
# def hkt(t):
#     exp = np.exp( t * evals )
#     lamb = np.diag(exp)
#     return U.dot( lamb  ).dot(U.T)

# def heat_kernel_signature(t_values):
#     all_results = []
#     for t in t_values:
#         kernel = hkt(t)
#         diagonal = np.diag(kernel)
#         rez = np.mean(diagonal)
#         all_results.append(rez)
#     return np.array(all_results)
```

```python
def hkt(t, e_vals, e_vecs):
    exp = np.exp( t * e_vals )
    lamb = np.diag(exp)
    return e_vecs.dot( lamb  ).dot(e_vecs.T)

def HKS(gr, t_vals):
    t_results = []
    lapl = -1*nx.laplacian_matrix(gr).toarray()

    e_vals, e_vecs = np.linalg.eigh(lapl)
    for t in t_vals:
        kernel = hkt(t, e_vals, e_vecs)
        diagonal = np.diag(kernel)
        rez = np.mean(diagonal)
        t_results.append(rez)
    return t_results
```

1b) Use scipy.spatial.cdist to compute a 60 x 60 matrix of distance values between these signatures. Look at the entries in this distance matrix. Are apples generally similar to apples? Are bones generally similar to bones?

```python
from scipy.spatial.distance import cdist
```

```python
sig_dists = cdist(hks_mat, hks_mat)
```

```python
np.diag(sig_dists).sum() == 0
```

```python
apple_idxs = [k for k in label_dict if label_dict[k] == 'apple']
beetle_idxs = [k for k in label_dict if label_dict[k] == 'beetle']
bone_idxs = [k for k in label_dict if label_dict[k] == 'bone']

grps_idx_list = [apple_idxs, beetle_idxs, bone_idxs]
```

```python

means = []
for l in grps_idx_list:
    
    total = 0
    for i in l:
        for j in l:
            if i != j: # choosing to ignore zero distances between node and self. I have done the same calculation with the sig_dist matrix to be fair
                total += sig_dists[i,j]
            
    means.append(total / ( (len(l) ** 2) - (len(l)) ) )

apple_mean_dst, beetle_mean_dst, bone_mean_dst = means
```

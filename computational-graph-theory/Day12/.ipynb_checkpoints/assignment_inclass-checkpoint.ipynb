```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linear_sum_assignment
from sklearn.neighbors import kneighbors_graph
```

```python
pts = np.random.random((100,2))
```

```python
plt.scatter(pts[:,0], pts[:,1])
```

```python
adjacency = kneighbors_graph(pts, 3)
gr = nx.from_scipy_sparse_matrix(adjacency)
```

```python
nx.draw(
    gr,
    pos = pts
)
```

```python
# want to find the best matching of these points such that length of all pairs is minimized
from scipy.spatial.distance import cdist
```

```python
cost_matrix = cdist(pts[:50],
                    pts[50:])
```

```python
cost_matrix.shape
```

```python
soln = linear_sum_assignment(cost_matrix)
```

```python
soln
```

```python
assignment_graph = nx.Graph()

for i in range(soln[0].shape[0]):
    assignment_graph.add_edge(soln[0][i], 50 + soln[1][i])
```

```python
nx.draw(
    assignment_graph,
    node_size=6
)
```

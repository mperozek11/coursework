0a) Implement Prim's Algorithm. You can assume that the input points you are given are 2D points in the plane. Your method should produce a list of edge which make up the minimum spanning tree of those points.



```python
# init dist(src) = 0
# dist(all other nodes) = inf
# put all nodes in pQ with p given by dist
# while pQ is not empty:
    # v = pQ.dequeue()
    # put v in the MST (connected to the node which it is closest to)
    # for each neighbor N of v:
        # if N is not in the MST:
            # new_dist = dist(v, N)
            # if new_dist < dist(N):
                # update dist(N) in pQ

def prims(gr):
    gr = gr.copy()
    src = list(gr.nodes())[0]
    pq = []
    
    MST = nx.Graph()
    # init dist(src) = 0
    # dist(all other nodes) = inf
    for node in gr:
        if node == src:
            gr.nodes[node]['dist'] = 0.0
            gr.nodes[node]['closest'] = None
            heapq.heappush(pq, (0.0, node))
        else:
            gr.nodes[node]['dist'] = np.inf
            gr.nodes[node]['closest'] = None
            heapq.heappush(pq, (np.inf, node))
            
    # while pQ is not empty:        
    while pq:
        v = heapq.heappop(pq)[1]
        MST.add_node(v)
        
        if gr.nodes[v]['closest'] != None:
            # add edge to closest node
            MST.add_edge(v, gr.nodes[v]['closest'])
        
        for neighbor in gr[v]:
            new_dst = gr[v][neighbor]['weight']

            if new_dst < gr.nodes[neighbor]['dist']:
                gr.nodes[neighbor]['dist'] = new_dst
                gr.nodes[neighbor]['closest'] = v
                heapq.heappush(pq, (new_dst, neighbor))
    return MST
```

```python
test_gr = nx.grid_graph((10,10))
```

```python
def euclid_dist(v1, v2):
    return np.linalg.norm(v1-v2)
```

```python
for node in test_gr:
    for neighbor in test_gr[node]:
        test_gr[node][neighbor]['weight'] = euclid_dist(np.array(node), np.array(neighbor))
```

```python
MST = prims(test_gr)
```

0b) Generate a couple of examples of random points in the plane, and use nx.draw to draw the minimum spanning tree of those points. Check if these results look like a reasonable MST.

```python
def connected_rand_gr(nodes): 
    pts = 10 * np.random.random((nodes,2))
    rand_gr = nx.erdos_renyi_graph(nodes, 1.0)
    for node in rand_gr:
        for neighbor in rand_gr[node]:
            rand_gr[node][neighbor]['weight'] = euclid_dist(pts[node,:], pts[neighbor,:])
    pos={i:pts[i,:] for i in rand_gr.nodes()}
    
    return rand_gr, pos
```

0e) Implement one of the greedy algorithms for TSP. This should be pretty easy, especially if you use a priority queue.

```python
# returns the weight of the edge between the two nodes in the graph
def dist(u,v,gr):
    return gr[u][v]['weight']

def path_2_edgelist(path):
    edges = []
    
    cur = path[0]
    for i in range(len(path)-1):
        edges.append((cur,path[i+1]))
        cur = path[i+1]
    edges.append((cur,path[0]))
    return edges

def greedy_tsp(gr, src=None):
    path_nodes = []
    if not src:
        src = list(gr.nodes())[0]
        
    path_nodes.append(src)
    nodes_to_visit = list(gr.nodes())
    nodes_to_visit.remove(src)
    cur = src
    while nodes_to_visit:
        pq = []
        for node in nodes_to_visit:
            heapq.heappush(pq, (dist(cur,node, gr), node))
        next_node = heapq.heappop(pq)[1]
        path_nodes.append(next_node)
        nodes_to_visit.remove(next_node)
        cur = next_node
    return path_2_edgelist(path_nodes)
        
```

1a) The above steps are all of the pieces you need to build Christofides' algorithm. You should use your own Prim's implementation, and use the scipy and networkx implementations of the other steps. Remember that the graph you construct needs to be a MultiGraph, not just a Graph. Also remember that the final part of the algorithm is discarding duplicate vertices.

```python
# 1st draft
def path_2_edgelist(path):
    edges = []
    
    cur = path[0]
    for i in range(len(path)-1):
        edges.append((cur,path[i+1]))
        cur = path[i+1]
    edges.append((cur,path[0]))
    return edges

def christo(gr):
    MST = prims(gr)
    path = []
    start = list(MST.nodes())[0]
    stack = []
    stack.append(start)

    while len(path) < len(MST.nodes()):
        cur = stack.pop()
        if not cur in path:
            for neighbor in MST[cur]:
                stack.append(neighbor)
            path.append(cur)
            
    TSP = nx.Graph()
    TSP.add_edges_from(path_2_edgelist(path))
    return TSP
    
```

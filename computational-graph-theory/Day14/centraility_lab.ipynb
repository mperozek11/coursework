```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
zach = nx.karate_club_graph()
```

```python
pos = nx.spectral_layout(zach)
```

```python
nx.draw(zach, pos=pos)
```

```python
color = nx.degree_centrality(zach)
```

```python
nx.draw(
    zach,
    pos=pos,
    node_color=[color[v] for v in list(zach.nodes)]
)
```

```python
color = nx.betweenness_centrality(zach)
```

```python
nx.draw(
    zach,
    pos=pos,
    node_color=[color[v] for v in list(zach.nodes)]
)
```

```python
color = nx.eigenvector_centrality(zach)
```

```python
nx.draw(
    zach,
    pos=pos,
    node_color=[color[v] for v in list(zach.nodes)]
)
```

```python
color = nx.closeness_centrality(zach)
```

```python
nx.draw(
    zach,
    pos=pos,
    node_color=[color[v] for v in list(zach.nodes)]
)
```

```python
# random graph social networks
```

```python
social = nx.barabasi_albert_graph(1000, 3)
```

```python
deg_list = [item[1] for item in social.degree()]
```

```python
unique = set(deg_list)
```

```python
pts_to_plot = []
```

```python
for degree in unique:
    pts_to_plot.append((degree, deg_list.count(degree)))
```

```python
pts_to_plot = np.array(pts_to_plot)
```

```python
plt.scatter(
    pts_to_plot[:,0],
    pts_to_plot[:,1]
)
plt.yscale('log')
plt.xscale('log')
```
